```python
import pandas as pd
```

```python
# df_d_cg = pd.read_csv('data120_1.csv')
# df_cg = pd.read_csv('data120_2.csv')
# df_d_dg = pd.read_csv('data120_3.csv')
# df_dg = pd.read_csv('data120_4.csv')
# df_d_wdg = pd.read_csv('data120_5.csv')
# df_wdg = pd.read_csv('data120_6.csv')
df_d_cg = pd.read_csv('data_disjoint_cg.csv')
df_cg = pd.read_csv('data_cg.csv')
df_d_dg = pd.read_csv('data_disjoint_dg.csv')
df_dg = pd.read_csv('data_dg.csv')
df_d_wdg = pd.read_csv('data_disjoint_wdg.csv')
df_wdg = pd.read_csv('data_wdg.csv')
datas = [df_d_cg, df_cg, df_d_dg, df_dg, df_d_wdg, df_wdg]
```

```python
for i in range(len(datas)):
    datas[i] = datas[i][datas[i]['expanded']!=0]
idx = datas[0].index
for i in range(len(datas)):
    idx = datas[i].index.intersection(idx)
```

```python
for i in range(len(datas)):
    datas[i] = datas[i].loc[idx]
df_solvable = pd.concat(datas)
df_solvable = df_solvable[df_solvable['density']==0.4]
df_solvable.to_csv('04_60_solvable.csv', index=False)
df_solvable['search_time']=df_solvable['cpu_time']-df_solvable['h_time']-df_solvable['mdd_time']
df_solvable['count']=1
```

```python
group = df_solvable.groupby(['heuristic', 'disjoint', 'num_agents','density'], as_index=False).sum()
for c in group.columns[4:-1]:
    group[c]=group[c]/group['count']
group.columns[5:-1]
```
